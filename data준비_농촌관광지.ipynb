```python
import numpy as np
```

```python
for col in ['체험','교육', '숙박', '음식']:
  print(col, df[col].unique())
```

```python
# 결측치 NA 값 넣기
for col in ['체험','교육', '숙박', '음식']:
  for idx in range(len(df.index)):
    if df[col][idx] == '1':
      df[col][idx] = 1
    elif df[col][idx] == '2' :
      df[col][idx] = 2
    elif df[col][idx] == '3':
      df[col][idx] = 3
    elif df[col][idx] == '  ' :
      df[col][idx] = np.nan
```

```python
df[['체험','교육', '숙박', '음식']] # 잘 들어갔나 확인
```

```python
# 컬럼별 값을 더하고 na를 제외한 값의 개수로 나눠 평균을 구했음
def NaMean(x):
  return x.sum() / x.count()
```

```python
df.groupby('시군구_변환')['체험','교육', '숙박', '음식'].agg(NaMean)
```

```python
dataframe2 = df.groupby('시군구_변환')['체험','교육', '숙박', '음식'].agg(NaMean)
dataframe2
```

```python
dataframe2 = dataframe2.fillna(df.mean()).round(3)
dataframe2
```

```python
df.groupby('시군구_변환')[['마을이름']].count() # 지역별 농촌마을 수
```
