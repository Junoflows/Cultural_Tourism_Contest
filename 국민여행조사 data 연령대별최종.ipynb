### 지역별로 각 컬럼별 평균 구하기

```python
# 컬럼별 값을 더하고 na를 제외한 값의 개수로 나눠 평균을 구했음
def NaMean(x):
  return x.sum() / x.count()
```

```python
# 컬럼별로 위 함수 적용
지역별_만족도20 = data_농촌20.groupby('지역')[data_농촌20.columns[1:-1]].agg(NaMean)
지역별_만족도20 = round(지역별_만족도20,2)
지역별_만족도20
```

```python
지역별_만족도30 = data_농촌30.groupby('지역')[data_농촌30.columns[1:-1]].agg(NaMean)
지역별_만족도30 = round(지역별_만족도30,2)
지역별_만족도40 = data_농촌40.groupby('지역')[data_농촌40.columns[1:-1]].agg(NaMean)
지역별_만족도40 = round(지역별_만족도40,2)
지역별_만족도50 = data_농촌50.groupby('지역')[data_농촌50.columns[1:-1]].agg(NaMean)
지역별_만족도50 = round(지역별_만족도50,2)
지역별_만족도60 = data_농촌60.groupby('지역')[data_농촌60.columns[1:-1]].agg(NaMean)
지역별_만족도60 = round(지역별_만족도60,2)
```
