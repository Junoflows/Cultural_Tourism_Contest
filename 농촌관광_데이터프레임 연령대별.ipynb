```python
base_df = pd.read_pickle('/content/drive/MyDrive/문화관광 빅데이터 공모전/대참사/농촌관광지df생성.pkl')
base_df
```

#### 신한카드_데이터 불러오기

```python
card_df = pd.read_pickle('/content/drive/MyDrive/문화관광 빅데이터 공모전/전처리/card_travel.pkl')
card_df
```

##### 2022년 데이터 추출

```python
card_df['TA_YM'].dt.year.value_counts()
```

```python
# 카드 데이터 2022년만 사용하기 (4681798개 -> 1698133개)
card_df_2022 = card_df[card_df['TA_YM'].dt.year == 2022].reset_index(drop=True)
card_df_2022
```

```python
card_df_2022.to_pickle('/content/drive/MyDrive/문화관광 빅데이터 공모전/대참사/card_df_2022.pkl')
```

##### 농촌지역에 해당하는 카드 사용 데이터 추출

```python
# 농촌관광지역만 추출 (1698133개 -> 969500개)
시군구_농촌 = base_df['관광지역_시군구'].values.tolist()
c_data_농촌 = card_df_2022[card_df_2022['가맹점시군구'].isin(시군구_농촌)]
```

```python
c_data_농촌 = c_data_농촌.reset_index(drop=True)
c_data_농촌
```

#### 20대 데이터프레임 생성

```python
df_age_20 = base_df.copy()
```

```python
# 20대 해당하는 농촌지역_카드 데이터 추출
c_data_20 = c_data_농촌[c_data_농촌['CLN_AGE_R'] == 20].reset_index(drop=True)
c_data_20
```

```python
c_data_20.to_pickle('/content/drive/MyDrive/문화관광 빅데이터 공모전/대참사/2022_농촌지역_카드_20대.pkl')
```

##### 카드 총 이용금액 & 이용건수

```python
c_data_20.groupby('가맹점시군구')[['VLM', 'USEC']].sum()
```

```python
df_age_20 = pd.merge(df_age_20, c_data_20.groupby('가맹점시군구')[['VLM', 'USEC']].sum(), how='left', left_on='관광지역_시군구', right_on='가맹점시군구')
```

```python
df_age_20.rename(columns={'VLM':'카드_총_이용금액', 'USEC':'카드_총_이용건수'}, inplace=True)
```

```python
df_age_20
```

##### 업종 소분류별_카드 이용건수

```python
c_data_20.groupby(['가맹점시군구','GB2'])[['USEC']].sum().unstack()
```

```python
df_age_20 = pd.merge(df_age_20, c_data_20.groupby(['가맹점시군구','GB2'])[['USEC']].sum().unstack()['USEC'], how='left', left_on='관광지역_시군구', right_on='가맹점시군구')
```

```python
df_age_20.rename(columns={'관광쇼핑':'카드_관광쇼핑_이용건수','교통':'카드_교통_이용건수', '숙박':'카드_숙박_이용건수', '여행사':'카드_여행사_이용건수', '체험':'카드_체험_이용건수'}, inplace=True)
```

##### 주중/주말 별 이용건수

```python
# 주중 / 주말 별 이용건수 (Red: 주말, White: 평일)
c_data_20.groupby(['가맹점시군구','DAW_CCD_R'])[['USEC']].sum().unstack()['USEC']
```

```python
df_age_20 = pd.merge(df_age_20, c_data_20.groupby(['가맹점시군구','DAW_CCD_R'])[['USEC']].sum().unstack()['USEC'], how='left', left_on='관광지역_시군구', right_on='가맹점시군구')
```

```python
df_age_20.rename(columns={'RED':'카드_주말_이용건수', 'WHITE':'카드_평일_이용건수'}, inplace=True)
```

```python
df_age_20
```

```python
df_age_20.to_pickle('/content/drive/MyDrive/문화관광 빅데이터 공모전/대참사/df_20대.pkl')
```
