## 만족도/재방문의사/타인추천의도 열 생성

```python
df = pd.read_pickle('/content/drive/MyDrive/문화관광 빅데이터 공모전/전처리/total_data.pkl')
df = df[(df['MON_EXP_1']=='예') | (df['MON_EXP_2']=='예')].reset_index(drop=True)
df
```

```python
recent_data = pd.DataFrame()
```

```python
# 만족도/재방문의사/타인추천의도 는 최근 여행으로만 조사됨
# 최근 여행 데이터프레임 생성
for idx in range(len(df)):
  if pd.isna(df['D_TRA6_1_SPOT'][idx]) == True :
    if pd.isna(df['D_TRA5_1_SPOT'][idx]) == True :
      if pd.isna(df['D_TRA4_1_SPOT'][idx]) == True :
        if pd.isna(df['D_TRA3_1_SPOT'][idx]) == True :
          if pd.isna(df['D_TRA2_1_SPOT'][idx]) == True :
            if pd.isna(df['D_TRA1_1_SPOT'][idx]) == True :
              continue
            else:
              recent_data.loc[idx,['여행형태']] = df.loc[idx,['D_TRA1_CASE']].tolist()
              recent_data.loc[idx,['지역']]  = df.loc[idx,['D_TRA1_1_SPOT']].tolist()
          else:
            recent_data.loc[idx,['여행형태']] = df.loc[idx,['D_TRA2_CASE']].tolist()
            recent_data.loc[idx,['지역']] = df.loc[idx,['D_TRA2_1_SPOT']].tolist()
        else:
          recent_data.loc[idx,['여행형태']] = df.loc[idx,['D_TRA3_CASE']].tolist()
          recent_data.loc[idx,['지역']] = df.loc[idx,['D_TRA3_1_SPOT']].tolist()
      else:
        recent_data.loc[idx,['여행형태']] = df.loc[idx,['D_TRA4_CASE']].tolist()
        recent_data.loc[idx,['지역']] = df.loc[idx,['D_TRA4_1_SPOT']].tolist()
    else:
      recent_data.loc[idx,['여행형태']] = df.loc[idx,['D_TRA5_CASE']].tolist()
      recent_data.loc[idx,['지역']] = df.loc[idx,['D_TRA5_1_SPOT']].tolist()
  else:
    recent_data.loc[idx,['여행형태']] = df.loc[idx,['D_TRA5_CASE']].tolist()
    recent_data.loc[idx,['지역']] = df.loc[idx,['D_TRA5_1_SPOT']].tolist()

recent_data
```

### 지역별로 평균

```python
# 컬럼별 값을 더하고 na를 제외한 값의 개수로 나눠 평균을 구했음
def NaMean(x):
  return x.sum() / x.count()
```

```python
# 컬럼별로 위 함수 적용
지역별_만족도 = data_농촌.groupby('지역')[data_농촌.columns[1:]].agg(NaMean)
지역별_만족도 = round(지역별_만족도,2)
지역별_만족도
```
